```python
import pandas as pd
import matplotlib.pyplot as plt

# File path to your extracted CSV
file_path = r"C:\Users\anime\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_38144\API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"

# Read the CSV (skip metadata rows from World Bank dataset)
df = pd.read_csv(file_path, skiprows=4)

# Choose year for visualization
year = "2022"
df_year = df[["Country Name", year]].dropna()

# Histogram of population distribution
plt.figure(figsize=(10, 6))
plt.hist(df_year[year], bins=30, color='skyblue', edgecolor='black')
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.title(f"Distribution of Population in {year}")
plt.show()

# Bar chart of top 10 countries by population
top_countries = df_year.sort_values(by=year, ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(top_countries["Country Name"], top_countries[year], color='orange')
plt.xticks(rotation=45, ha='right')
plt.ylabel("Population")
plt.title(f"Top 10 Countries by Population in {year}")
plt.show()
```
