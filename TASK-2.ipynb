```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

titanic = pd.read_csv(r"C:\Users\anime\Downloads\titanic.csv")

titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])
titanic['fare'] = titanic['fare'].fillna(titanic['fare'].median())

print(titanic.info())

# Plot 1: Count of Survivors
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='survived', palette='pastel')
plt.title("Survivor Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.tight_layout()
plt.show()

# Plot 2: Survival by Gender
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=titanic, x='sex', hue='survived', palette='Set2')
plt.title("Survival by Gender")
plt.xlabel("Sex")
plt.ylabel("Count")
ax.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()

# Plot 3: Age Distribution
plt.figure(figsize=(8, 4))
sns.histplot(titanic['age'], bins=30, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Plot 4: Survival by Passenger Class
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=titanic, x='pclass', hue='survived', palette='Set1')
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
ax.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```
